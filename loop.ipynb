```python
s = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    phone = input("Enter phone = ")
    s = s+name+"--"+phone+"\n"
    
print(s)
```

```python
# # 4! = 1*2*3*4
# i ->range(1,4+1) = 1,2,3,4
fac = 1
n = int(input("Enter n = "))
for i in range(1,n+1):
    fac = fac*i
    
print(fac)
```

```python
a = "Hello World"
for i in a:
    print(i)
```

```python
a = "Hello World"
for i in a:
    print(i,end = "")
```

```python
a = "Hello World I am Python"
for i in a:
    if i != " ":
        print(i,end = "")
```

```python
for i in range(10):
    if i ==5:
        break
    else:
        print(i)
```

```python
for i in range(10):
    if i ==5:
        continue
        
    print(i)
```

```python
a = "Hello World I am Python"
for i in a:
    if i == " ":
        continue
    print(i,end = "")
```

```python
for i in range(1,51):
    if i%5 != 0:
        continue
        
    print(i)
```

```python
i = 0
s = 0
n = int(input("Enter n = "))
while i<n:
    x = int(input("Enter x = "))
    s = s+x
    i+=1
print(s)
```

```python
i = 0
s = str()
n = int(input("Enter n = "))
while i<n:
    name = input("Enter name = ")
    phone = input("Enter phone = ")
    s = s+name+"--"+phone+"\n"
    i = i+1
print(s)
```
