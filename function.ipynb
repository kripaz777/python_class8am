```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif (o == "/") and (b !=0):
        print(div(a,b))
    elif (o == "/") and (b == 0):
        print("The value of b can not be zero.")
    else:
        print("Invalid operator")
        
cal()
```

```python
def signup():
    print("-------------------Please signup------------------")
    username = input("Enter username = ")
    password = input("Enter password = ")
    cpassword = input("Enter cpassword = ")
    if cpassword == password:
        print("You are signedup !")
        return username,password
        
    else:
        print("password and confirm password is not same")
        signup()
        
username,password = signup()

def auth():
    print("--------------Login-----------------")
    x = input("Enter username = ")
    y = input("Enter password = ")
    
    if x != username and y != password:
        auth()
    else:
        print("You are logedin")
        
auth()
```
